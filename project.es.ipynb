#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('assets/real_estate.csv', sep=';')
poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
subconjunto = df[df['level5'].isin(poblaciones)]
subconjunto['price_normalized'] = subconjunto.groupby('level5')['price'].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
fig, ax = plt.subplots()
for poblacion in poblaciones:
    subset = subconjunto[subconjunto['level5'] == poblacion]
    subset['price_normalized'].plot(kind='hist', bins=20, alpha=0.5, ax=ax, label=poblacion)
plt.title('Histogramas Normalizados de Precios por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
